```python
x_axis = list(range(1,100))
```

```python
seq = [(2*n-3)/(3*n+4) for n in x_axis]
```

```python
fig = plt.plot(x_axis, seq)
plt.hlines(2/3,x_axis[0],x_axis[-1], color='red')
plt.show()
```

$\{a_n\} = \frac{2n - 3}{3n + 4} \rightarrow \frac{2}{3}, n \rightarrow \infty$

$a_n = \frac{\sqrt{n}}{n+2}$

```python
seq_2 = [(n**(1/2))/(n+2) for n in x_axis]
```

```python
fig2 = plt.plot(x_axis[:50], seq_2[:50])
# plt.hlines(2/3,x_axis[0],x_axis[-1], color='red')
plt.show()
```

```python
seq = [0,1] # a_1 = 1
x_axis = list(range(1,100))
for i in x_axis:
    seq.append(3 - 1/seq[i]) # a_{n+1} = 3 - 1/a_n
fig2 = plt.plot(x_axis, seq[1:-1])
# plt.hlines(2/3,x_axis[0],x_axis[-1], color='red')
plt.show()
```

```python
seq = [0,2]
x_axis = list(range(1,10))
for i in x_axis:
    seq.append(1/(3-seq[i]))
fig2 = plt.plot(x_axis, seq[1:-1])
plt.show()
for an in seq[1:-1]:
    print("{} = {}".format(Fraction(an), an)) 
```

Equação:  
$
\left\{ 
\begin{array}{c}
a_1 = \sqrt{2} \\
a_{n+1} = \sqrt{2+a_n}
\end{array}
\right. 
$

```python
seq = [0,sqrt(2)]
x_axis = list(range(1,10))
for i in x_axis:
    seq.append(sqrt(2 + seq[i]))
fig2 = plt.plot(x_axis, seq[1:-1])
plt.show()
```

```python
seq = [0,1]
x_axis = list(range(1,10))
for i in x_axis:
    seq.append(1+(1/(1+seq[i])))
fig2 = plt.plot(x_axis, seq[1:-1])
plt.show()
```
